## BLOQUE USUARIO

Vamos a relacionar las variables categóricas de VIP group y User Status con la variable a predecir de RISK (sabemos que este bloque de usuario condiciona en gran medida esta variable a predecir)

````python
pd.crosstab(df.VIP_group, df.User_status)
````

````python
vip_group_order = ['Ceo', 'Directiva', 'Ciber', 'Otros']  # Reemplaza con tus categorías y orden deseado
df['VIP_group'] = pd.Categorical(df['VIP_group'], categories=vip_group_order, ordered=True)

pd.crosstab(df.VIP_group, df.RISK).plot(kind="bar")
````

Se pueden sacar las conclusiones de que los CEO y directiva suelen tener un reisgo muy alto en su mayoría, los ciber high y los demas entre medium y low, lo cual es de esperar.

Los valores que "no encajan", por ejemplo que salgan low en CEO hacen posiblemente referencia a usuarios inactivos, lo comprobamos con otra gráfica:

````python
vip_group_order = ['Ceo', 'Directiva', 'Ciber', 'Otros']  # Reemplaza con tus categorías y orden deseado
risk_order = ['Very high', 'High', 'Medium','Low', 'Very low']  # Reemplaza con tus categorías y orden deseado

# Convertir las columnas a categóricas con el orden especificado
df['VIP_group'] = pd.Categorical(df['VIP_group'], categories=vip_group_order, ordered=True)
df['RISK'] = pd.Categorical(df['RISK'], categories=risk_order, ordered=True)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
fig.subplots_adjust(wspace=0.3)

pd.crosstab(df.RISK, df.User_status).plot(kind="bar",ax=ax[0])
pd.crosstab(df.VIP_group, df.User_status).plot(kind="bar", ax=ax[1])
ax[0].set_title('RISK vs User Status')
ax[1].set_title('VIP Group vs User Status')
````

### Conclusiones BLOQUE USERS:
Efectivamente vemos con la primera gráfica que tanto en very high como en high todos los usuarios son activos y en very low son inactivos. Esto indica que el hecho de que un usuario se activo o inactivo lleva el resultado final RISK a los extremos

Los valores intermedios como medium y low tienen credenciales de usuarios activos e inactivos, sin embargo vemos que en medium hay más activos y en low más inactivos (lo cual corrobora la conlusi´pon anterior)

Respecto a la segunda gráfica vemos que en todos grupos hay algunos usuarios inactivos aunque en menor cantidad, lo cual nos sirve para explicar por qué por ejemplo en el grupo CEO (que asociamos con un alto riesgo) a veces salían resultados de riesgo medio o bajo según el diagrama de la celda anterior

### Matriz de correlación
Podemos visualizar de manera conjunta las relaciones entre todos los posibles pares de variables y seleccionar los que resulten más relevantes

````python
# Dejamos solo las columnas de tipo numerico significativas
df_nums=df.copy()

df_nums=df_nums.select_dtypes(include='number')
df_nums=df_nums.drop(columns=['Muestra_ID'])

coef=df_nums.corr(method='pearson')

plt.figure(figsize=(10,6))
sns.heatmap(coef, annot=True, vmin=-1, vmax=1, cmap='coolwarm', center=0)
plt.show()
````

Tras representar la matriz de correlación de todas las variables NUMÉRICAS del dataframe, vemos que no existen correlaciones significativas entre variables de distintos bloques pero sí existen correlaciones significativas entre variables del mismo bloque interesantes de estudiar. Por ello, haremos matrices de correlación para estudiar aquellos bloques que tienen muchas variables de tipo numérico: PASSWORD y CHANNEL


````python
# Listado de índices de correlación en sentido descendente (índices negativo indican variables que tienen relación inversa)
corr = df_nums.corr()
corr[['TOTAL']].sort_values(by = 'TOTAL',ascending = False).style.background_gradient()
````

#### Bloque Password y Bloque Channel

````python
df_password=df_nums[['Leaked_password', 'Guesses_discover', 'Cracking_time', 'Password_strength', 'Password_entropy']].copy()
df_channel=df_nums[['Subscribers', 'Engagement_rate', 'Mentions', 'Posts_day', 'Reposts']].copy()
coef1=df_password.corr(method='pearson')
coef2=df_channel.corr(method='pearson')
plt.figure(figsize=(10,6))
ax = sns.heatmap(coef1, annot=True, vmin=-1, vmax=1, cmap='coolwarm', center=0)
ax.set_title('Matrix Passwords')
````

````python
plt.figure(figsize=(10,6))
ax = sns.heatmap(coef2, annot=True, vmin=-1, vmax=1, cmap='coolwarm', center=0)
ax.set_title('Matrix Channels')
````

Observaciones:
* Bloque password: Las variables guesses to discover, password entropy y password strength tienen una alta correlación
* Bloque channel: Hay una elevada correlación entre los suscriptores de un canal y el número de reenvíos de las publicaciones del canal. También entre los reenvíos y el engagement. Entre los suscriptores y el engagement.

Estas observaciones no aportan información respecto a la variable a predecir del riesgo de las credenciales, pero nos dan información del comportamiento de las variables que se están estudiando entre ellas.
